## 1 Histogram : We should use histogram when we need the count of the variable in a plot.

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#national park not numpy
np_data=pd.read_csv('national_parks.csv')
```

```python
np_data.head()
```

```python
np_data.info()
```

```python
plt.hist(np_data['GrandCanyon'], facecolor='lightgreen', bins=10, edgecolor='blue')
plt.show
```

### Components of a histogram
* <b>n:</b> Contains the frequency of each bin
* <b>bins:</b> Represents the middle value of each bin
* <b>patches:</b> The Patch object for the rectangle shape representing each bar

```python
n, bins, patches= plt.hist(np_data['GrandCanyon'], 
                           facecolor='lightgreen', 
                           bins=10, 
                           edgecolor='blue')
print('n: ', n)
print('bins: ', bins)
print('patches: ', patches)
```

#### The cumulative property
If True, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values. The last bin gives the total number of datapoints.

```python
plt.hist(np_data['GrandCanyon'], 
        facecolor='peru', 
        bins=10, 
        edgecolor='blue',
        cumulative=True)
plt.show()
```

#### Check the histogram to a range of values
We only look at the data points within the range 2M-5M. This realigns the bins in the histogram

```python
plt.hist(np_data['GrandCanyon'], 
        facecolor='peru', 
        bins=10, 
        edgecolor='blue',
        range=(2000000,5000000))
plt.show()
```

## Multiple Histogram

```python
plt.hist(np_data['GrandCanyon'], 
        facecolor='peru', 
        bins=10, 
        edgecolor='blue',
        alpha=0.3)
plt.hist(np_data['BryceCanyon'], 
        facecolor='orangered',
        edgecolor='maroon',
        bins=10)
plt.show()
```

## 2  Pie Chart: Pie chart shows distribution of data based on proportion of pie occupied. 

#### Importing Libraries and Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

movie_data=pd.read_csv('types_movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.info()
```

```python
plt.pie(movie_data['Percentage'],
       labels=movie_data['Sector'])
plt.axis('equal')
plt.show
```

### Pie chart components
* <b>wedges:</b> A list of Patch objects representing each wedge
* <b>texts:</b> List of Text objects representing the labels
* <b>autotexts:</b> List of Text objects for the numeric values - this is only availabe if the autopct value for the pie chart is not None

```python
wedges, texts, autotexts=plt.pie(movie_data['Percentage'],
                                labels=movie_data['Sector'],
                                autopct='%.1f')
plt.axis('equal')
print('wedges: ', wedges)
print('texts: ', texts)
print('autotexts: ', autotexts)
```

#### Pie chart customizations
 We set the values for the colors and autopct properties. The latter sets the format for the values to be displayed. 

https://matplotlib.org/stable/gallery/color/named_colors.html

```python
colors_list=['darkorange','sandybrown','darksalmon','orangered','chocolate']
```

```python
plt.pie(movie_data['Percentage'], 
        labels=movie_data['Sector'],
       colors=colors_list, 
       autopct='%.2f')
plt.axis('equal')
plt.show()
```

#### The explode property
To highlight a particular wedge of the pie chart, we use explode to separate it from the rest of the chart. 

The value for "explode" represents the fraction of the radius with which to offset each wedge.

```python
colors_list=['darkorange','sandybrown','darksalmon','orangered','chocolate']
explode=(0, 0.1, 0,0,0.3)

plt.pie(movie_data['Percentage'], 
        labels=movie_data['Sector'],
       colors=colors_list, 
       autopct='%.2f',
       explode=explode)
        
plt.axis('equal')
plt.show()
```

## 4 Boxplot 
Boxplot gives a nice summary of the data. It helps in understanding our distribution better.

* The end of the box shows the upper quartile(75%)and the start of the box represents the lower quartile(25%). 
* Upper Quartile is also called 3rd quartile and similarly, Lower Quartile is also called as 1st quartile. 
* The region between lower quartile and the upper quartile is called as Inter Quartile Range(IQR) and it is used to approximate the 50% spread in the middle data(75â€“25=50%).

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
exam_data=pd.read_csv('exams.csv')
```

```python
exam_data.head()
```

```python
exam_data.columns
```

```python
exam_scores=exam_data[['math score', 'reading score', 'writing score']]
```

```python
exam_scores.head()
```

```python
exam_scores.describe()
```

###  Box plots

```python
color=['blue', 'red','green']
bp=plt.boxplot(exam_scores, 
               patch_artist=True,
              notch=True)
plt.xticks([1,2,3], ['Math', 'Reading', 'Writing'])
for patch, color in zip(bp['boxes'],color):
    patch.set_facecolor(color)

plt.show()
```

### Violin Plots
Violin plot is a better chart than boxplot as it gives a much broader understanding of the distribution. It resembles a violin and dense areas point the more distribution of data otherwise hidden by box plots


```python
vp=plt.violinplot(exam_scores, showmedians=True)
plt.xticks([1,2,3],['Math', 'Reading', 'Writing'])
plt.show()
```

### Plotting two lines with 2 separate Y axes with the same X axis
First create one axis and then use the twinx() function to create another axis sharing the same X axis


```python


fig=plt.figure(figsize=(12,6))

# x axis
ax1.set_xlabel('Date')
# disable ticks, and label bottom disable x axis label
ax1.tick_params(axis='x', bottom=False, labelbottom=False)

# set left y axis
ax1.set_ylabel('Temp(F)', color='red', size='x-large')

ax1.tick_params(axis='y', labelcolor='red', labelsize='large')

ax1.plot(austin_weather['Date'], austin_weather['TempAvgF'],color='red')

#set right y axis

#twinxsets the same x axis for both plots

ax2=ax1.twinx()
ax2.set_ylabel('Average wind speed (MPH)', color='blue', size='x-large')

ax2.tick_params(axis='y', labelcolor='blue', labelsize='large')

ax2.plot(austin_weather['Date'], austin_weather['WindAvgMPH'],color='blue')

plt.show()
```

## 6 Stack Plot 
* Stack plot visualizes data in stacks and shows the distribution of data over time.
* It is used for checking multiple variable area plots in a single plot.

```python
np_data=pd.read_csv('national_parks.csv')
```

```python
np_data.head()
```

```python
x=np_data['Year']
y=np.vstack([np_data['Badlands'], 
             np_data['GrandCanyon'],
            np_data['BryceCanyon']])
```

```python
labels=['Badlands', 'GrandCanyon', 'BryceCanyon']
colors=['sandybrown', 'tomato', 'skyblue']
plt.stackplot(x,y,labels=labels, colors=colors, edgecolor='black')
plt.legend(loc=2)
```

```python
plt.figure(figsize=(10,7))
data_perc=np_data.divide(np_data.sum(axis=1), axis=0)
plt.stackplot(x, data_perc['Badlands'], 
              data_perc["GrandCanyon"], 
              data_perc['BryceCanyon'], 
              edgecolor='black', 
             colors=colors, 
             labels=labels)
plt.legend(loc='center left', bbox_to_anchor=(1,.05))
plt.show()
```

## 7 Bar Plot
* Bar Plot shows the distribution of data over several groups. 
* It is commonly confused with a histogram which only takes numerical data for plotting. 
* It helps in comparing multiple numeric values.

```python
countries=pd.read_csv('countries.csv')
```

```python
countries.head()
```

```python
countries.info()
```

```python
data_2007=countries[countries['year']==2007]
data_2007.info()
```

```python
x=range(10)
plt.figure(figsize=(12,6))
top_10_p=data_2007.sort_values('population', ascending=False). head(10)
plt.xticks(x, top_10_p['country'])
plt.bar(x, top_10_p['population']/10**9)
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('Ten Most Populated Countries in 2007')
plt.show()
```

## 8 Scatter Plot
* Scatter plot helps in visualizing 2 numeric variables. 
* It helps in identifying the relationship of the data with each variable i.e correlation or trend patterns. 
* It also helps in detecting outliers in the plot.
* It is used in Machine learning concepts like regression, where x and y are continuous variables. It is also used in clustering scatters or outlier detection.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.scatter(data_2007['gdpPerCapita'], data_2007['lifeExpectancy'])
plt.xlabel('GDP per Capita ($)')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs GDP per capita in the year 2007')
plt.show()
```
